```python
import nltk
from nltk import download as nltk_download

nltk_download('punkt')

import tkinter as tk
from textblob import TextBlob
from newspaper import Article

def summarize():
    url = utext.get('1.0', 'end')
    article = Article(url)
    article.download()
    article.parse()
    article.nlp()

    title.config(state='normal')
    authors.config(state='normal')
    publish_date.config(state='normal')
    summary.config(state='normal')
    sentiment.config(state='normal')

    title.delete('1.0', 'end')
    title.insert('1.0', article.title)

    authors.delete('1.0', 'end')
    authors.insert('1.0', article.authors)

    publish_date.delete('1.0', 'end')
    if article.publish_date:
        publish_date.insert('1.0', article.publish_date)
    else:
        publish_date.insert('1.0', "Publication date not available")

    summary.delete('1.0', 'end')
    summary.insert('1.0', article.text)

    analysis = TextBlob(article.text)

    sentiment.delete('1.0', 'end')
    sentiment.insert(
        '1.0', f'Polarity: {analysis.polarity}, Sentiment: {"Positive" if analysis.polarity > 0 else "Negative" if analysis.polarity < 0 else "Neutral"}'
    )

    title.config(state='disabled')
    authors.config(state='disabled')
    publish_date.config(state='disabled')
    summary.config(state='disabled')
    sentiment.config(state='disabled')

root = tk.Tk()
root.title("News Analysis 0.1")
root.geometry('600x700')

tlabel = tk.Label(root, text="Title",font='Calibri')
tlabel.pack()
title = tk.Text(root, height=1, width=140)
title.config(state='disabled', bg='#ADD8E6')
title.pack()

alabel = tk.Label(root, text="Author",font='Calibri')
alabel.pack()
authors = tk.Text(root, height=1, width=140)
authors.config(state='disabled', bg='#ADD8E6')
authors.pack()

dlabel = tk.Label(root, text="Date",font='Calibri')
dlabel.pack()
publish_date = tk.Text(root, height=1, width=140)
publish_date.config(state='disabled', bg='#ADD8E6')
publish_date.pack()

slabel = tk.Label(root, text="Summary",font='Calibri')
slabel.pack()
summary = tk.Text(root, height=20, width=140)
summary.config(state='disabled', bg='#ADD8E6')
summary.pack()

stlabel = tk.Label(root, text="Sentiment", font='Calibri')
stlabel.pack()
sentiment = tk.Text(root, height=1, width=140)
sentiment.config(state='disabled', bg='#ADD8E6')
sentiment.pack()

utlabel = tk.Label(root, text="URL",font='Calibri')
utlabel.pack()
utext = tk.Text(root, height=1, width=140)
utext.pack()

btn = tk.Button(root, text="Summarize",font='Calibri', command=summarize)
btn.pack()

root.mainloop()
```
